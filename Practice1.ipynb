Теперь отсортируем массив *X* с помощью **сортировки выбором**, которую я реализовал написаной мною функцией **selectSort**, которая принимает массив в виде аргумента, сортирует его, а затем возвращает.

Затем снова построим график по отсортированному *Х* и прежнему *Y*

Код и результат — график с точками красного цвета: 


````python
import numpy as nu
import random
import matplotlib.pyplot as plt
X=nu.random.randint(-20, 20, 30)
Y=nu.random.randint(-20, 20, 30)
def selectSort (arr):
    size=len(arr)
    for i in range(size):
        index=i
        value=arr[i]
        j=i+1
        while(j<size):
            if arr[j]<value:
                index=j
                value=arr[j]
            j+=1
        arr[index]=arr[i]
        arr[i]=value
    return arr
X=selectSort(X)        
plt.plot(X, Y, 'ro')
plt.show()
````
