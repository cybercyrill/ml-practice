Создаем таблицу на основе файла:

```python
import graphlab
sf=graphlab.SFrame('people-example.csv')
sf
```

Рандомизируем числа для столбца номер дома с помощью numpy.random.randint()

```python
import numpy as nu
sf['Номер дома'] = nu.random.randint(1, 150, 56)
sf
```

Создаем столбец ФИО из столбцов Имя и Фамилия:

```python
sf['ФИО'] = sf['Имя'] + ' ' + sf['Фамилия']
sf
```

Логическим фильтром отбираем необходимы значения для двух условий и выводим количество полученных записей:

```python
mem = sf[(sf['ФИО'] == 'Семен Семенов') | (sf['Страна'] == 'США')]
mem.head(len(mem))
```

```python
print(len(mem))
```

```python
memm = sf[(sf['ФИО'] == 'Семен Семенов') | (sf['Страна'] == 'США')& (sf['Возраст'] != 24)]
memm.head(len(memm))
```
